```python
import random
from four_letter_words import words
random_word = random.choice(words)
LR = list(random_word)

print("Welcome to WORDLE!")
print("GOAL: To find the target word in 5 attempts")
print("Instructions:")
print("(1) GREY, if letter is not present in the target word")
print("(2) YELLOW, if letter is present in the word but not in the right spot")
print("(3) GREEN, if letter is present at the correct position of the target word")
print("(4) Enter 4 letter words ONLY\n")
print("Let's begin")

attempts = 5
for guess_number in range(attempts):
    input1 = input(f"\nEnter your guess ({attempts - guess_number} attempts remaining): ")
    
    if len(input1) != 4:
        print("Invalid word length. Please enter a 4-letter word.")
        continue 
    elif input1 not in words:
        print("Invalid word. Please try again with another word.")
        continue

    L = []
    for i in range(4):
        if input1[i] == LR[i]:
            L.append("GREEN")
        elif input1[i] in LR and input1[i] != LR[i]:
            L.append("YELLOW")
        else:
            L.append("GREY")
            
    print("Guess",guess_number + 1,":",input1,"-",L,)
    if input1 == random_word:
        print("CONGRATS! You found the target word in",guess_number + 1,"guesses!")
        break

if guess_number == attempts - 1:
    print("\nSorry, you ran out of guesses. The word was:",random_word)
```
